```python
from transformers import AutoModelForCausalLM

model_id = "microsoft/Phi-3-vision-128k-instruct"

# Load the model and return the outputs of intermediate layers
model = AutoModelForCausalLM.from_pretrained(model_id,
                                             cache_dir="/content/my_models/phi_3_vision",
                                             device_map="cuda",
                                             trust_remote_code=True,
                                             torch_dtype="auto",
                                             _attn_implementation="eager",
                                             output_hidden_states=True)  # Ensure hidden states are returned
```

```python
from transformers import AutoProcessor
processor = AutoProcessor.from_pretrained(model_id, trust_remote_code=True)

messages = [
    {"role": "user", "content": "<|image_1|>\n1.)What all you can observe from the image? 2.)Is it daylight? 3.)Which color shirt is the person wearing beside blue car ?"}
]

prompt = processor.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

# Load and display the image
import matplotlib.pyplot as plt
from PIL import Image
import requests

url = "https://d1oudmv8ilx665.cloudfront.net/newroads/wp-content/uploads/2019/02/01075051/city-driving.jpg"
image = Image.open(requests.get(url, stream=True).raw)
plt.imshow(image)
plt.axis('off')
plt.show()

# Prepare the inputs for the model
inputs = processor(prompt, [image], return_tensors="pt").to("cuda:0")
```

```python
import torch
```

```python
# Forward pass through the model to get latent representation
with torch.no_grad():
    outputs = model(**inputs)

# Extract the latent representation from hidden states
# Access the second-to-last layer for embeddings
latent_representation = outputs.hidden_states[-2]

# Check the shape of the latent representation
print(latent_representation.shape)  # For example, (batch_size, seq_length, hidden_size)
```

```python
# Average the latent representation across the sequence length (2400) to get a single vector per image
latent_representation_mean = latent_representation.mean(dim=1)  # Shape: (1, 3072)

# Now, you have a 3072-dimensional embedding for the image
print(latent_representation)  # Output: torch.Size([1, 3072])
```
